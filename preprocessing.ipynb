```python
import pandas as pd

# Read the CSV files
covid_df = pd.read_csv('datasets/covid_yearly.csv')
economic_df = pd.read_csv('datasets/data_2020_2023.csv')

# Display initial info
print(f"COVID dataset shape: {covid_df.shape}")
print(f"Economic dataset shape: {economic_df.shape}")

# Perform inner join to keep only matching countries and years
merged_df = pd.merge(
    covid_df,
    economic_df,
    left_on=['Country', 'Year'],
    right_on=['country_name', 'year'],
    how='inner'
)

# Drop duplicate columns (country_name and year since we have Country and Year)
merged_df = merged_df.drop(['country_name', 'year'], axis=1)

# Display merge results
print(f"\nMerged dataset shape: {merged_df.shape}")
print(f"\nFirst few rows of merged dataset:")
print(merged_df.head())

# Save the merged dataset
merged_df.to_csv('datasets/merged.csv', index=False)
print("\nMerged dataset saved to 'datasets/merged.csv'")

# Show which countries are in the final dataset
merged_df.head(10)
```

```python
# Load the CSV, treating empty strings as NaN
df = pd.read_csv("datasets/merged.csv", na_values=["", " "])

# Drop rows where 'GDP Growth (% Annual)' is missing
df = df.dropna(subset=["GDP Growth (% Annual)"])

# Overwrite the original CSV
df.to_csv("datasets/merged.csv", index=False)
```
